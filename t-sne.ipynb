```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import fetch_openml

features, targets = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
```

```python
example = features[0]

n = int(np.sqrt(example.size))

example = example.reshape((n,n));

plt.imshow(example,cmap='binary')
```

```python
first_class = features[np.where(targets=='1')]

second_class = features[np.where(targets=='9')]
```

```python
second_class.shape
```

```python
two_classes = np.concatenate((first_class, second_class), axis=0)
two_classes.shape
```

```python
from sklearn.manifold import TSNE

features_embedded = TSNE(n_components=2, learning_rate='auto',
                  init='random', perplexity=3).fit_transform(two_classes)
```

```python
plt.scatter(features_embedded[:,0],features_embedded[:,1], alpha=0.5, color=['red' if targets[i]=='1' else  'blue' for i in range(len((two_classes)))])
```
