```python
import matplotlib.pyplot as plt

def plot_2d_histogram(features, bins=50):
    fig, ax = plt.subplots(ncols=2, figsize=(10,4), sharey=True)
    data_initial = features[0,:,:]
    data_final = features[-1,:,:]
    x0, y0 = data_initial[:, 0], data_initial[:, 1]
    xf, yf = data_final[:, 0], data_final[:, 1]
    h0 = ax[0].hist2d(x0, y0, bins=bins, density=True, cmap='inferno')
    h1 = ax[1].hist2d(xf, yf, bins=bins, density=True, cmap='inferno')
    ax[0].set_xlabel(r'$x$')
    ax[0].set_ylabel(r'$y$')
    ax[0].set_title(r'$s = T$')
    ax[1].set_xlabel(r'$x$')
    ax[1].set_title(r'$s = 0$')
    plt.colorbar(h0[3], ax=ax[0], label='Density')
    plt.colorbar(h1[3], ax=ax[1], label='Density')

    #plt.tight_layout()
    plt.savefig('figuras/gaussian_mixture.pdf')
    plt.show()
```

```python
distros = TrainAndGenerateDatasets(ndata=10000,dimension=2)
```

```python
plot_2d_histogram(distros)
```
